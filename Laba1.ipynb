```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Загрузка и обработка изображения
image = cv2.imread('test.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
sharpened = cv2.filter2D(image, -1, kernel)
sharpened_2 = cv2.addWeighted(image, 1.5, blurred_image, -0.5, 0)
edges = cv2.Sobel(image, cv2.CV_64F, 1, 1, ksize=5)
edges = cv2.convertScaleAbs(edges)
combined = cv2.addWeighted(blurred_image, 0.8, edges, 0.2, 0)
combined_2 = cv2.addWeighted(combined, 0.5, sharpened, 0.5, 0)

# Создание подграфиков
plt.figure(figsize=(20, 5))

# Отображение оригинального изображения
plt.subplot(2, 3, 1)
plt.imshow(image)
plt.axis('off')

# Отображение размытого изображения
plt.subplot(2, 3, 2)
plt.imshow(blurred_image)
plt.axis('off')

# Отображение sharpened изображения
plt.subplot(2, 3, 3)
plt.imshow(sharpened)
plt.axis('off')

# Отображение sharpened_2 изображения
plt.subplot(2, 3, 4)
plt.imshow(sharpened_2)
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(edges)
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(combined)
plt.axis('off')

# Показать все изображения
plt.tight_layout()
plt.show()



```
